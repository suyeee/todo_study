# 지역 별 최저 온도 찾기
- 1800.csv 파일을 불러오면 됩니다.
- `지역코드, 날짜, 최고/최저여부,온도, ...`
- `TMAX` : 최고 온도
- `TMIN` : 최저 온도
- 파이썬 내장 함수중 `min()` 사용

```python
from pyspark import SparkConf, SparkContext
conf = SparkConf().setMaster('local').setAppName('region-temperature')
sc = SparkContext(conf=conf)
```

```python
filepath = '/home/lab26/SparkCourse/data/1800.csv'
```

```python
lines = sc.textFile(f"file:///{filepath}")
lines.collect()
```

```python
# 각 행을 파싱하기 위한 함수
def parse(row):
    fields = row.split(",")
    
    # 지역이름 가져오기
    region = fields[0]
    
    # 온도 가져오기
    temp = int(fields[3])
    
    return region, temp
```

```python
rows = lines.map(parse).collect()
rows
```

```python
lst1=[]
lst2=[]
lst3=[]
for r, t in rows:
    if r == 'ITE00100554':
        lst1.append(t)
        
    elif r== 'GM000010962':
        lst2.append(t)
    else:
        lst3.append(t)
lst1
```

```python
min(lst1)
```
